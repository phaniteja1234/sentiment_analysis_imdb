```python
def clean_html(txt):
    clean=re.compile("<.*?>")
    return re.sub(clean,'',txt)
```

```python
df['review']=df['review'].apply(clean_html)
```

```python
def clean_tx(txt):
    pat=re.compile(r"[>]+")
    return re.sub(pat,'',txt)
```

```python
df['review']=df['review'].apply(clean_tx)
```

```python
## converting everything to lower
def textlower(txt):
    return txt.lower()
```

```python
df['review']=df['review'].apply(textlower)
```

```python
## removing special chars
```

```python
def remove(txt):
    x=''
    for i in txt:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
df['review']=df['review'].apply(remove)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words("english")
```

```python
def stopwrds(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words("english"):
            x.append(i)
    y=x[:]
    x.clear()
    return y
            
```

```python
df['review']=df['review'].apply(stopwrds)
```

```python
## perform stemming
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
pr=PorterStemmer()
```

```python
pr.stem('running')
```

```python
def stemming(txt):
    y=[]
    for i in txt:
        y.append(pr.stem(i))
    return y
```

```python
df['review']=df['review'].apply(stemming)
```
